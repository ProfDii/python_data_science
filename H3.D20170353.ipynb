```python
# This function returns the nth value of a fibonacci series
def fibonacci(n):
    '''This function returns the nth number in a fibonacci series.
    It does this by accepting an argument that is an interger.
    It will return an error if the argument isn't an integer.
    '''
    
    # check if the argument in the function is an integer
    if not isinstance (n, int):
        print ("this is not an integer")
    
    # the default first element in a fibonacci series is 0
    #it isn't dependent on addition of previous series element
    elif n == 0:
        return 0
    
    # the second element of a fibonacci series
    elif n == 1:
        return 1
    
    # we are making use of a recurssive function
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
# This program generates the first 25 numbers of a fibonacci series

# First, we declare all the required variables
Number = 25
old = 0
new = 1

# we set a loop to repeat itself 25 times
for x in range(0, Number):
    
    # the first numbers of a fibonacci series is 0 and 1
    # so we take care of those first
    if(x <= 1):
        Next = x
    
    # to generate the remaining numbers...
    else:
        
        # we add the 2 previous numbers
        Next = old + new
        
        # we reassign variables for the next iteration
        old = new
        new = Next
        
    # we print the number
    print(Next)
```

```python
# function to get unique values 

def delete_duplicates(list1):
    '''This function takes a list as argument and returns the list without deuplicates
    '''
    
    # we declare our list without duplicates as an empty list
    list_without_duplicates = []
    
    # we go through all the values in our list one-by-one
    for x in list1:
        
        # we check if the each value in our list already exists in list_without_duplicates
        if x not in list_without_duplicates:
            
            # if it isn't, we append it to list_without_duplicates
            list_without_duplicates.append(x) 
    
    # we print out the list without duplicates
    for x in list_without_duplicates: 
        print (x)
```
